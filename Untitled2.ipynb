```python
data_country_wise=pd.read_csv(r'D:\Sudha\InSaid_Project\Emission_countrywise.csv')
print('data_country_wise')

data_top10_countries=pd.read_csv(r'D:\Sudha\InSaid_Project\Emissions_top_10_countries.csv')
print('Emissions_top_10_countries')

data_methane_global=pd.read_csv(r'D:\Sudha\InSaid_Project\Methane_project.csv')
print('data_methane_global')
```

```python
#plotting global methane levels
#data from https://gml.noaa.gov/ccgg/trends_ch4/
#plt.figure(figsize=(60, 40))
sns.set(rc = {'figure.figsize':(15,8)})
sns.set_style("whitegrid")
x=data_methane_global.year
y=data_methane_global.average_ppb
z=data_methane_global.trend
plt.plot(data_methane_global.year, data_methane_global.average_ppb, color='blue')
sns.lineplot(x, y, label='Methane Global Average ppb' )
sns.lineplot(x, z, label='trend', color='red')
plt.title(['Monthly Methane global average from 1983 to 2021'])

#plt.show()
plt.savefig('Methane_global_average.png')
```

```python
pd.crosstab(index=data_top10_countries['country'], columns=[data_top10_countries['co2'], data_top10_countries['population']], margins=True).style.background_gradient(cmap='winter')
```

```python
fig = plt.figure(figsize=(15, 7))

# Creating countplot of country with methane_per_Capita
ax = sns.countplot(x='co2_per_capita', hue='methane_per_capita', data=data_top10_countries, palette='RdBu')

# Adding percentages to the bars
total = data_top10_countries.shape[0]
for p in ax.patches:
    percentage = '{:.2f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width()/5
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y))
    
# Plot Design
#plt.xticks(ticks=[0, 1], labels=["co2_per_capita", "methane_per_capita"], fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(xlabel="co2_per_capita", size=14)
plt.ylabel(ylabel='methane_per_capita', size=14)
plt.title(label='Comparison of co2_per_capita and methane_per_capita', size=16)
plt.legend(labels=["co2_per_capita", "methane_per_capita"])
plt.grid(b=True)

# Display the figure
plt.show()
```
